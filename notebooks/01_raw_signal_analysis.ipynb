Lowpass filter experiment

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy.signal import butter, filtfilt

limit_freq = 2000

# 1. Low-pass filter do 5000 Hz
def lowpass_filter(signal, fs, cutoff=limit_freq, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    filtered = filtfilt(b, a, signal)
    return filtered

signal_filtered = lowpass_filter(signal_32, sample_rate)

# 2. CWT
scales = np.arange(1, 200)  # dopasuj skal do fs i pasma
coefficients, frequencies = pywt.cwt(signal_filtered, scales, 'morl', sampling_period=1/sample_rate)

# 3. Ograniczenie widoku do 0-5000 Hz
freq_limit = limit_freq
idx_limit = np.where(frequencies <= freq_limit)[0]

plt.figure(figsize=(12, 6))
plt.imshow(np.abs(coefficients[idx_limit, :]), aspect='auto',
           extent=[0, len(signal_32)/sample_rate, frequencies[idx_limit[-1]], frequencies[idx_limit[0]]],
           vmax=np.max(np.abs(coefficients[idx_limit, :])))

plt.yscale('linear')
plt.colorbar(label='Amplitude (absolute)')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.title(f'CWT (filtered, up to {limit_freq} Hz)')
plt.show()
```

Time domain

```python
rows = []

for index, sample in main_df.iterrows():
    signal = sample["signal"]
    if signal.shape == (0,):
        print(f"Signal for file {sample['filename']} is 0 shaped")
        continue
    feats = extract_time_features(signal)

    feats["temperature_set"] = sample["temperature_set"]
    feats["temperature"] = sample["temperature"]

    rows.append(feats)
```

```python
df = pd.DataFrame(rows)
print(df.head())
```

```python
sns.boxplot(data=df, x="temperature_set", y="rms")
plt.title("RMS vs temperature")
plt.show()
sns.boxplot(data=df, x="temperature_set", y="std")
plt.title("STD vs temperature")
plt.show()
sns.boxplot(data=df, x="temperature_set", y="ptp")
plt.title("Peak-to-Peak vs temperature")
plt.show()
sns.boxplot(data=df, x="temperature_set", y="energy")
plt.title("Energy vs temperature")
plt.show()
sns.boxplot(data=df, x="temperature_set", y="zcr")
plt.title("Zero crossing rate vs temperature")
plt.show()
sns.boxplot(data=df, x="temperature_set", y="crest_factor")
plt.title("Crest factor vs temperature")
plt.show()
```

```python
from scipy.stats import kruskal

groups = [df[df.temperature_set == t]["rms"] for t in sorted(df.temperature_set.unique())]
stat, p_rms = kruskal(*groups)

groups = [df[df.temperature_set == t]["std"] for t in sorted(df.temperature_set.unique())]
stat, p_std = kruskal(*groups)

groups = [df[df.temperature_set == t]["ptp"] for t in sorted(df.temperature_set.unique())]
stat, p_ptp = kruskal(*groups)

groups = [df[df.temperature_set == t]["energy"] for t in sorted(df.temperature_set.unique())]
stat, p_en = kruskal(*groups)

groups = [df[df.temperature_set == t]["zcr"] for t in sorted(df.temperature_set.unique())]
stat, p_zcr = kruskal(*groups)

groups = [df[df.temperature_set == t]["crest_factor"] for t in sorted(df.temperature_set.unique())]
stat, p_cf = kruskal(*groups)

results = {
    "Feature": ["RMS", "STD", "PTP", "Energy", "ZCR", "Crest Factor"],
    "p-value": [p_rms, p_std, p_ptp, p_en, p_zcr, p_cf]
}

df_stats = pd.DataFrame(results)
df_sorted = df_stats.reindex(df_stats['p-value'].abs().sort_values(ascending=True).index)
df_sorted
```

```python
from scipy.stats import spearmanr

rho_rms, p_rms = spearmanr(df["temperature"], df["rms"])
rho_std, p_std = spearmanr(df["temperature"], df["std"])
rho_ptp, p_ptp = spearmanr(df["temperature"], df["ptp"])
rho_en, p_en = spearmanr(df["temperature"], df["energy"])
rho_zcr, p_zcr = spearmanr(df["temperature"], df["zcr"])
rho_cf, p_cf = spearmanr(df["temperature"], df["crest_factor"])

results = {
    "Feature": ["RMS", "STD", "PTP", "Energy", "ZCR", "Crest Factor"],
    "Correlation": [rho_rms, rho_std, rho_ptp, rho_en, rho_zcr, rho_cf],
    "Correlation p-value": [p_rms, p_std, p_ptp, p_en, p_zcr, p_cf]
}

df_stats = pd.DataFrame(results)
df_sorted = df_stats.reindex(df_stats['Correlation'].abs().sort_values(ascending=False).index)
df_sorted
```

```python
from sklearn.linear_model import LinearRegression 
X = df[["temperature"]] 
y = df["rms"] 
model = LinearRegression().fit(X, y)

plt.scatter(df["temperature"], df["rms"], alpha=0.5, label="Samples")

x_line = np.linspace(df["temperature"].min(),
                     df["temperature"].max(), 100)
y_line = model.predict(x_line.reshape(-1, 1))

plt.plot(x_line, y_line, color="red", label="Linear regression")
plt.xlabel("Temperature [C]")
plt.ylabel("RMS")
plt.legend()
plt.title("RMS vs temperature")
plt.show()
```

```python
bands = ["fft_energy_low", "fft_energy_mid", "fft_energy_high"]

corr_bands = df[bands].corr(method="spearman")
print(corr_bands)
```

```python
import pandas as pd
from scipy.stats import spearmanr

results = []

for col in df.columns:
    if col == "temperature":
        continue
    if col.endswith("Hz_energy"):
        continue

    rho, p = spearmanr(df["temperature"], df[col])
    results.append({
        "feature": col,
        "spearman_rho": rho,
        "p_value": p
    })

corr_df = pd.DataFrame(results)
corr_df = corr_df.sort_values(by="spearman_rho", key=lambda x: abs(x), ascending=False)

pd.set_option("display.max_rows", None)
print(corr_df)
```

```python
pd.set_option("display.max_rows", None)

corr_matrix = df.corr(method="spearman")
corr_temp = corr_matrix["temperature"]
filtered = corr_temp[~corr_temp.index.str.endswith("Hz_energy")]

print(filtered.sort_values(ascending=False))
```

```python
fft_columns = [col for col in df.columns if "Hz_energy" in col]

corrs = df[fft_columns + ["temperature"]].corr(method="spearman")["temperature"].drop("temperature")

corr_df = corrs.reset_index()
corr_df.columns = ["band", "correlation"]

corr_df["start_freq"] = corr_df["band"].str.extract(r"(\d+)-")[0].astype(int)
corr_df = corr_df.sort_values("start_freq")
```

```python
plt.figure(figsize=(12,5))
plt.bar(corr_df["start_freq"], corr_df["correlation"], width=100, color="skyblue")
plt.xticks(rotation=90)
plt.xlabel("Pasmo czstotliwoci [Hz]")
plt.ylabel("Korelacja z temperatur")
plt.title("Korelacja energii pasm FFT z temperatur")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
